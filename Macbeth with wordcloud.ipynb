```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from wordcloud import WordCloud
```

```python
path = "/content/macbetty.txt"

myFile = open(path, "r")

myStr = re.sub(r"[^a-zA-Z0-9']", ' ', myFile.read().casefold())

data = myStr.split()

s = pd.Series(data)

output = s.value_counts().to_dict()

for key, value in output.items():
  [print(key + ':', value*'*')]
```

```python
s = s.value_counts()
s[:10].plot(kind='bar')
plt.show()
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(myStr)
# Display the generated Word Cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
