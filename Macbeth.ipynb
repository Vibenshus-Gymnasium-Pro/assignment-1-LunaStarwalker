```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
path = "/content/macbetty.txt"

myFile = open(path, "r")

myStr = re.sub(r'[^a-zA-Z0-9]', ' ', myFile.read().casefold())

data = myStr.split()

s = pd.Series(data)

output = s.value_counts().to_dict()

[print(key + ':', value*'*') for key, value in output.items()]
```

```python
s = s.value_counts()
s[:10].plot(kind='bar')
plt.show()
```
